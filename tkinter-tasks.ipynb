### Simple GUI calculator using Tkinter

```python
from tkinter import *

class calculator:
    expression = ""
    
    def __init__(self,root):
        self.root = root
        self.root.title("Calculator")
        self.root.geometry("312x324")
        self.root.resizable(0, 0)
        self.input_text = StringVar()
        self.expression = ''
        self.expression_field = Entry(self.root, textvariable=self.input_text)
        self.expression_field.grid(columnspan=4, ipadx=70)
        self.button1 = Button(root, text=' 1 ', fg='black', bg='grey',command=lambda: self.press(1), height=1, width=7)
        self.button1.grid(row=2, column=0)
        self.button2 = Button(root, text=' 2 ', fg='black', bg='grey',command=lambda: self.press(2), height=1, width=7)
        self.button2.grid(row=2, column=1)
        self.button3 = Button(root, text=' 3 ', fg='black', bg='grey',command=lambda: self.press(3), height=1, width=7)
        self.button3.grid(row=2, column=2)
        self.button4 = Button(root, text=' 4 ', fg='black', bg='grey',command=lambda: self.press(4), height=1, width=7)
        self.button4.grid(row=3, column=0)
        self.button5 = Button(root, text=' 5 ', fg='black', bg='grey',command=lambda: self.press(5), height=1, width=7)
        self.button5.grid(row=3, column=1)        
        self.button6 = Button(root, text=' 6 ', fg='black', bg='grey',command=lambda: self.press(6), height=1, width=7)
        self.button6.grid(row=3, column=2) 
        self.button7 = Button(root, text=' 7 ', fg='black', bg='grey',command=lambda: self.press(7), height=1, width=7)
        self.button7.grid(row=4, column=0) 
        self.button8 = Button(root, text=' 8 ', fg='black', bg='grey',command=lambda: self.press(8), height=1, width=7)
        self.button8.grid(row=4, column=1)
        self.button9 = Button(root, text=' 9 ', fg='black', bg='grey',command=lambda: self.press(9), height=1, width=7)
        self.button9.grid(row=4, column=2)
        self.button0 = Button(root, text=' 0 ', fg='black', bg='grey',command=lambda: self.press(0), height=1, width=7)
        self.button0.grid(row=5, column=0)
        self.plus = Button(root, text=' + ', fg='black', bg='grey',command=lambda: self.press("+"), height=1, width=7)
        self.plus.grid(row=2, column=3)
        self.minus = Button(root, text=' - ', fg='black', bg='grey',command=lambda: self.press("-"), height=1, width=7)
        self.minus.grid(row=3, column=3)
        self.multiply = Button(root, text=' * ', fg='black', bg='grey',command=lambda: self.press("*"), height=1, width=7)
        self.multiply.grid(row=4, column=3)
        self.divide = Button(root, text=' / ', fg='black', bg='grey',command=lambda: self.press("/"), height=1, width=7)
        self.divide.grid(row=5, column=3)
        self.equal = Button(root, text=' = ', fg='black', bg='grey',command=self.equalpress, height=1, width=7)
        self.equal.grid(row=5, column=2)
        self.clear = Button(root, text='Clear', fg='black', bg='grey',command=self.clear, height=1, width=7)
        self.clear.grid(row=5, column='1')
        self.Decimal= Button(root, text='.', fg='black', bg='grey',command=lambda: self.press('.'), height=1, width=7)
        self.Decimal.grid(row=6, column=0)        
        
    def press(self,num):
        global expression
        self.expression = self.expression + str(num)
        self.input_text.set(self.expression) 

    def equalpress(self):
        try:
            global expression
            total = str(eval(self.expression))
            self.input_text.set(total)
            self.expression = ""
        except:
 
            self.input_text.set(" error ")
            self.expression = ""
    def clear(self):
        global expression
        self.expression = ""
        self.input_text.set("")
 
```

```python
root = Tk()
c = calculator(root)
root.mainloop()
```

### File Explorer in Python using Tkinter

```python
from tkinter import *
from tkinter import filedialog

class file_explorer:
    
    def __init__(self,root):
        self.root = root
        self.root.title("File Explorer")
        self.root.geometry("500x500")
        self.root.config(bg='white')
        self.label1 = Label(self.root,text = "File Explorer using Tkinter",width = 100, height = 4,fg = "black")
        self.label1.grid(row=1, column=1)
        self.button_exp = Button(self.root,text = "Browse Files",command = self.browseFiles)
        self.button_exp.grid(row=2, column=1)       
        
    def browseFiles(self):
        filename = filedialog.askopenfilename(initialdir = "/",
                                          title = "Select a File",
                                          filetypes = (("Text files",
                                                        "*.txt*"),
                                                       ("all files",
                                                        "*.*")))
      
        self.label1.configure(text="File Opened: "+filename)
```

```python
root = Tk()
c = file_explorer(root)
root.mainloop()
```

### Weight Conversion GUI using Tkinter

```python
from tkinter import *
from tkinter import messagebox

class weight_conversion:
    
    
    def __init__(self,root):
        self.root = root
        self.root.title("Weight Conversion")
        self.root.geometry("600x300")
        self.root.config(bg='white')
        self.clicked = StringVar()
        self.clicked.set( "KG" )
        self.clicked1 = StringVar()
        self.clicked1.set( "G" )
        self.options = ["KG","G","Pound","Ounce" ]
        self.label1 = Label(self.root,text = "Input the Weight",width = 20, height = 4)
        self.label1.grid(row=1, column=1)
        self.entry1 = Entry(self.root)
        self.entry1.grid(row=1, column=2) 
        self.label2 = Label(self.root,text = "From Unit",width = 20, height = 4)
        self.label2.grid(row=2, column=1)
        self.drop = OptionMenu( self.root , self.clicked , *self.options )
        self.drop.grid(row=2, column=2)
        self.label3 = Label(self.root,text = "To Unit",width = 20, height = 4)
        self.label3.grid(row=2, column=3)
        self.drop1 = OptionMenu( self.root , self.clicked1 , *self.options )
        self.drop1.grid(row=2, column=4)  
        self.label4 = Label(self.root,text = "Converted Weight",width = 20, height = 4)
        self.label4.grid(row=3, column=1)        
        self.t1 = Entry(self.root)
        self.t1.grid(row = 3, column=2)
        self.button_exp = Button(self.root,text = "Convert",command = self.convert)
        self.button_exp.grid(row=4, column=1)       
        
    def convert(self):
        option1 = self.clicked.get()
        option2 = self.clicked1.get()
        if option1 == option2:
            messagebox.showinfo("info","From unit and To unit are same. Please change")
        elif option1 == 'KG' and option2 == 'G':
            kg_to_g = float(self.entry1.get())*1000
            self.t1.delete(0,END)
            self.t1.insert(0,kg_to_g)
        elif option1 == 'KG' and option2 == 'Pound':
            kg_to_p = float(self.entry1.get())*2.20462
            self.t1.delete(0,END)            
            self.t1.insert(0,kg_to_p)
        elif option1 == 'KG' and option2 == 'Ounce':
            kg_to_o = float(self.entry1.get())*35.274
            self.t1.delete(0,END)            
            self.t1.insert(0,kg_to_o)  
        elif option1 == 'G' and option2 == 'KG':
            g_to_kg = float(self.entry1.get())/1000
            self.t1.delete(0,END)            
            self.t1.insert(0,g_to_kg)  
        elif option1 == 'G' and option2 == 'Pound':
            g_to_p = float(self.entry1.get())/454
            self.t1.delete(0,END)            
            self.t1.insert(0,g_to_p)    
        elif option1 == 'G' and option2 == 'Ounce':
            g_to_o = float(self.entry1.get())/28.35
            self.t1.delete(0,END)            
            self.t1.insert(0,g_to_o)  
        elif option1 == 'Pound' and option2 == 'KG':
            p_to_kg = float(self.entry1.get())/2.20462
            self.t1.delete(0,END)            
            self.t1.insert(0,p_to_kg)   
        elif option1 == 'Pound' and option2 == 'G':
            p_to_g = float(self.entry1.get())*454
            self.t1.delete(0,END)            
            self.t1.insert(0,p_to_g)              
        elif option1 == 'Pound' and option2 == 'Ounce':
            p_to_o = float(self.entry1.get())*16
            self.t1.delete(0,END)            
            self.t1.insert(0,p_to_o)      
        elif option1 == 'Ounce' and option2 == 'KG':
            o_to_kg = float(self.entry1.get())/35.274
            self.t1.delete(0,END)            
            self.t1.insert(0,o_to_kg)   
        elif option1 == 'Ounce' and option2 == 'G':
            o_to_g = float(self.entry1.get())*28.35
            self.t1.delete(0,END)            
            self.t1.insert(0,o_to_g)              
        elif option1 == 'Ounce' and option2 == 'Pound':
            o_to_p = float(self.entry1.get())/16
            self.t1.delete(0,END)            
            self.t1.insert(0,o_to_p)              
```

```python
root = Tk()
c = weight_conversion(root)
root.mainloop()
```

### Age Calculator using Tkinter

```python
from tkinter import *
from tkinter import messagebox
from datetime import date

class age_calculator:
    
    
    def __init__(self,root):
        self.root = root
        self.root.title("Age Calculator")
        self.root.geometry("400x300")
        self.root.config(bg='Light blue')
        self.name = Label(text="Name").grid(row=1, column=0, padx=90)
        self.year = Label(text="Year").grid(row=2, column=0)
        self.month = Label(text="Month").grid(row=3, column=0)
        self.day = Label(text="Day").grid(row=4, column=0)
        self.nameValue = StringVar()
        self.yearValue = StringVar()
        self.monthValue = StringVar()
        self.dayValue = StringVar()

        self.nameEntry = Entry(self.root, textvariable=self.nameValue)
        self.nameEntry.grid(row=1, column=1, pady=10)
        self.yearEntry = Entry(self.root, textvariable=self.yearValue)
        self.yearEntry.grid(row=2, column=1, pady=10)
        self.monthEntry = Entry(self.root, textvariable=self.monthValue)
        self.monthEntry.grid(row=3, column=1, pady=10)
        self.dayEntry = Entry(self.root, textvariable=self.dayValue)
        self.dayEntry.grid(row=4, column=1, pady=10)     

        self.button_calc = Button(self.root,text = "Calculate Age",command = self.calculate_age)
        self.button_calc.grid(row=5, column=1)       
        
    def calculate_age(self):
        try:
            today = date.today()
            birthDate = date(int(self.yearEntry.get()), int(self.monthEntry.get()), int(self.dayEntry.get()))
            if today < birthDate:
                messagebox.showinfo("info","Cannot calculate age for future period")           
            else:
                age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day))
                Label(text=f"{self.nameValue.get()} your age is {age} years").grid(row=6, column=1)
        except Exception as e:
            messagebox.showwarning("Error!", str(e))
```

```python
root = Tk()
c = age_calculator(root)
root.mainloop()
```

### Create a digital clock using Tkinter

```python
from tkinter import *
from tkinter import messagebox
import time

class digitalclock:
    
    
    def __init__(self,root):
        self.root = root
        self.root.title("Digital Clock")
        self.root.geometry("300x150")
        self.root.config(bg='grey')
        self.root.resizable(0,0)
        self.label1 = Label(self.root,bg = '#962317',fg='#f0f0f0', bd = 60,font=("Courier", 30, 'bold'))
        self.label1.grid(row=0, column=1) 
        
    def digital_clock(self):
        try:
            time_input = time.strftime("%H:%M:%S")
            self.label1.config(text=time_input)
            self.label1.after(200, self.digital_clock)
        except Exception as e:
            messagebox.showwarning("Error!", str(e))
```

```python
root = Tk()
c = digitalclock(root)
c.digital_clock()
root.mainloop()
```

### Make Notepad using Tkinter

```python
import re
from tkinter import *
from tkinter.ttk import *
from datetime import datetime
from tkinter import messagebox
from tkinter import filedialog,simpledialog
from tkinter.scrolledtext import ScrolledText

class mynotepad:
    
    
    def __init__(self,root):
        self.root = root
        self.root.title("Vignesh Notepad")
        self.root.geometry("900x900")
        self.root.config(bg='grey')
        self.notepad = ScrolledText(self.root, width = 90, height = 40)
        self.filename = ''
        self.notepadMenu = Menu(self.root)
        self.root.configure(menu=self.notepadMenu)
        #file menu
        self.fileMenu = Menu(self.notepadMenu, tearoff = False)
        self.notepadMenu.add_cascade(label='File', menu = self.fileMenu)
        self.fileMenu.add_command(label='New', command = self.cmdNew)
        self.fileMenu.add_command(label='Open', command = self.cmdOpen)
        self.fileMenu.add_command(label='Save', command = self.cmdSave)
        self.fileMenu.add_command(label='Save As', command = self.cmdSaveAs)
        self.fileMenu.add_separator()
        self.fileMenu.add_command(label='Exit', command = self.cmdExit)    
        self.notepad.pack()
        #edit menu
        self.editMenu = Menu(self.notepadMenu, tearoff = False)
        self.notepadMenu.add_cascade(label='Edit', menu = self.editMenu)
        self.editMenu.add_command(label='Cut', command = self.cmdCut)
        self.editMenu.add_command(label='Copy', command = self.cmdCopy)
        self.editMenu.add_command(label='Paste', command = self.cmdPaste)
        self.editMenu.add_command(label='Delete', command = self.cmdClear)
        self.editMenu.add_separator()
        self.editMenu.add_command(label='Find', command = self.cmdFind)
        self.editMenu.add_command(label='Select All', command = self.cmdSelectAll)
        #help menu
        self.helpMenu = Menu(self.notepadMenu, tearoff = False)
        self.notepadMenu.add_cascade(label='Help', menu = self.helpMenu)
        self.helpMenu.add_command(label='About Notepad', command = self.cmdAbout)  

        
    def cmdNew(self):
        if len(self.notepad.get('1.0', END+'-1c'))>0:
            if messagebox.askyesno("Notepad", "Do you want to save changes?"):
                self.cmdSave()
            else:
                self.notepad.delete(0.0, END)
                self.root.title("Notepad")
                
    def cmdOpen(self):
        fd = filedialog.askopenfile(parent = self.root, mode = 'r')
        t = fd.read()    
        self.notepad.delete(0.0, END)
        self.notepad.insert(0.0, t)       
        
    def cmdSave(self):
        fd = filedialog.asksaveasfile(mode = 'w', defaultextension = '.txt')
        if fd!= None:
            data = self.notepad.get('1.0', END)
        try:
            fd.write(data)
        except Exception as e:
            messagebox.showerror(title="Error", message = "Unable to save the file! due to {}".format(str(e)))
        
    def cmdSaveAs(self):
        fd = filedialog.asksaveasfile(mode='w', defaultextension = '.txt')
        t = self.notepad.get(0.0, END)     
        try:
            fd.write(t.rstrip())
        except Exception as e:
            messagebox.showerror(title="Error", message = "Unable to save file! due to {}".format(str(e)))
            
    def cmdExit(self):
        if messagebox.askyesno("Notepad", "Are you sure you want to exit?"):
            self.root.destroy()

    def cmdCut(self):     
        self.notepad.event_generate("<<Cut>>")
    def cmdCopy(self):     
        self.notepad.event_generate("<<Copy>>")
    def cmdPaste(self):     
        self.notepad.event_generate("<<Paste>>")
    def cmdClear(self):     
        self.notepad.event_generate("<<Clear>>")
       
    def cmdFind(self):    
        self.notepad.tag_remove("Found",'1.0', END)
        find = simpledialog.askstring("Find", "Find what:")
        if find:
            idx = '1.0'    
        while 1:
            idx = self.notepad.search(find, idx, nocase = 1, stopindex = END)
            if not idx:
                break
            lastidx = '%s+%dc' %(idx, len(find))
            self.notepad.tag_add('Found', idx, lastidx)
            idx = lastidx
            self.notepad.tag_config('Found', foreground = 'black', background = 'yellow')
            self.notepad.bind("<1>", self.click)
            
    def click(self,event):     
        self.notepad.tag_config('Found',background='white',foreground='black')
        
    def cmdSelectAll(self):    
        self.notepad.event_generate("<<SelectAll>>")
            
    def cmdAbout(self):    
        self.label = messagebox.showinfo("About Notepad", "Notepad by - Vignesh as a part of the task given by iNeuron")
        
```

```python
root = Tk()
a = mynotepad(root)
root.mainloop()
```

### How to create a COVID19 Data Representation GUI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tkinter as tk
from tkinter import *
from tkinter.ttk import *
from datetime import datetime
from tkinter import messagebox
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, NavigationToolbar2Tk)


class covidanalysis:
    
    
    def __init__(self,root):
        self.root = root
        self.root.title("COVID-19 Data Representation")
        self.root.geometry("1500x1500")
        self.root.config(bg='#04AFD3')
        self.label1 = Label(self.root,background = '#04AFD3', text = "***************************************************")
        self.label1.grid(row=0, column=1)         
        self.label2 = Label(self.root,background = '#04AFD3',font=("Courier", 20, 'bold'), text = "COVID-19 Data Representation")
        self.label2.grid(row=0, column=2) 
        self.label3 = Label(self.root,background = '#04AFD3', text = "***************************************************")
        self.label3.grid(row=0, column=3) 
        self.df = self.read_data('country_wise_latest.csv')
        self.region_uni = self.df['WHO Region'].unique()
        self.label4 = Label(self.root,background = '#04AFD3',font=("Courier", 14, 'bold'), text = "Select the Region")
        self.label4.grid(row=1, column=1)  
        self.region = Combobox( self.root , width=37, value= tuple(self.region_uni)) 
        self.region.grid(row=1, column=2,padx=10, pady=2, sticky='w')   
        
        self.label5 = Label(self.root,background = '#04AFD3',font=("Courier", 14, 'bold'), text = "Select the Country")
        self.label5.grid(row=2, column=1)       
        self.Country = Combobox( self.root , width=37)
        self.Country.grid(row=2, column=2,columnspan=2, padx=10, pady=2, sticky='w')  
        self.Country.bind('<Button-1>', self.callback)   
        self.button_show = Button(self.root,text = "Show Stats",command = self.show_stats)
        self.button_show.grid(row=3, column=1)         
       
        
    def read_data(self,filepath):
        df = pd.read_csv(filepath)
        return df
    
    def callback(self,eventObject):
        abc = eventObject.widget.get()
        region = self.region.get()
        self.country_uni = self.df[self.df['WHO Region'] == region]['Country/Region'].unique()
        self.Country.config(values=tuple(self.country_uni))
    
    def show_stats(self):
        country = self.Country.get()
        conf = self.df[self.df['Country/Region']==country]['Confirmed'].iloc[0]
        deaths = self.df[self.df['Country/Region']==country]['Deaths'].iloc[0]
        recovered = self.df[self.df['Country/Region']==country]['Recovered'].iloc[0]
        new_cases = self.df[self.df['Country/Region']==country]['New cases'].iloc[0]
        new_deaths = self.df[self.df['Country/Region']==country]['New deaths'].iloc[0]
        new_recovered = self.df[self.df['Country/Region']==country]['New recovered'].iloc[0]        
        self.label6 = Label(self.root,background = '#04AFD3',font=("Courier", 14, 'bold'), text = "Confirmed")
        self.label6.grid(row=4, column=1)
        self.ent1 = Entry(self.root)
        self.ent1.grid(row=4,column=2)
        self.ent1.delete(0,END)            
        self.ent1.insert(0,conf)        
        self.label7 = Label(self.root,background = '#04AFD3',font=("Courier", 14, 'bold'), text = "Deaths")
        self.label7.grid(row=5, column=1)        
        self.ent2 = Entry(self.root)
        self.ent2.grid(row=5,column=2) 
        self.ent2.delete(0,END)            
        self.ent2.insert(0,deaths)           
        self.label8 = Label(self.root,background = '#04AFD3',font=("Courier", 14, 'bold'), text = "Recovered")
        self.label8.grid(row=6, column=1)         
        self.ent3 = Entry(self.root)
        self.ent3.grid(row=6,column=2)  
        self.ent3.delete(0,END)            
        self.ent3.insert(0,recovered) 
        ###########################################
        self.label9 = Label(self.root,background = '#04AFD3',font=("Courier", 14, 'bold'), text = "New Cases")
        self.label9.grid(row=4, column=3)
        self.ent4 = Entry(self.root)
        self.ent4.grid(row=4,column=4)
        self.ent4.delete(0,END)            
        self.ent4.insert(0,new_cases)        
        self.label10 = Label(self.root,background = '#04AFD3',font=("Courier", 14, 'bold'), text = "New Deaths")
        self.label10.grid(row=5, column=3)        
        self.ent5 = Entry(self.root)
        self.ent5.grid(row=5,column=4) 
        self.ent5.delete(0,END)            
        self.ent5.insert(0,new_deaths)           
        self.label11 = Label(self.root,background = '#04AFD3',font=("Courier", 14, 'bold'), text = "New Recovered")
        self.label11.grid(row=6, column=3)         
        self.ent6 = Entry(self.root)
        self.ent6.grid(row=6,column=4)  
        self.ent6.delete(0,END)            
        self.ent6.insert(0,new_recovered) 
        self.button_vis = Button(self.root,text = "Visualize",command = self.visualize)
        self.button_vis.grid(row=7, column=1)          

    def visualize(self):
        region = self.region.get()
        country = self.Country.get()
        figure1 = plt.Figure(figsize=(4,4), dpi = 100)
        ax1 = figure1.add_subplot(111)
        bar1 = FigureCanvasTkAgg(figure1, self.root)
        bar1.get_tk_widget().grid(row=9,column=1)
        covid_df[(covid_df['WHO Region']==region) & (covid_df['Country/Region']==country)][['Confirmed','Deaths','Recovered']].plot(kind='bar',legend=True,ax=ax1)
        ax1.set_title('Total Confirmed/deaths/recovered')
        figure2 = plt.Figure(figsize=(4,4), dpi = 100)
        ax2 = figure2.add_subplot(111)
        bar2 = FigureCanvasTkAgg(figure2, self.root)
        bar2.get_tk_widget().grid(row=9,column=2)        
        covid_df[(covid_df['WHO Region']==region) & (covid_df['Country/Region']==country)][['New cases','New deaths','New recovered']].plot(kind='bar',legend=True,ax=ax2)
        ax2.set_title('New cases/Deaths/recovered')        
        
```
